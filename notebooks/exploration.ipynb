```python
import pandas as pd

df = pd.read_csv('../data/DataCoSupplyChainDataset.csv', encoding='latin-1')

# exploring the data
print(df.shape)
```

```python
# fill empty values with unknown for customer lname
df['Customer Lname'] = df['Customer Lname'].fillna('Unknown')

# fill empty Customer Zipcode with 00000
df['Customer Zipcode'] = df['Customer Zipcode'].fillna(0)
```

```python
# drop columns Product Description, Customer Password and Order Zipcode
df = df.drop(columns=['Product Description', 'Customer Password', 'Order Zipcode', 'Customer Email'])
```

```python
# print the number of columns in the dataset
print(len(df.columns))
```

```python
# check for empty values again after cleaning
print("Missing values", df.isnull().sum().sum())
```

```python
# convert Order Date, Shipping Date to datetime format
df['order date (DateOrders)'] = pd.to_datetime(df['order date (DateOrders)'])
df['shipping date (DateOrders)'] = pd.to_datetime(df['shipping date (DateOrders)'])

print(df[['order date (DateOrders)', 'shipping date (DateOrders)']].dtypes)
```

```python
print("Shape:", df.shape)
print("Missing values:", df.isnull().sum().sum())
print("Date columns:")
print(df[['order date (DateOrders)', 'shipping date (DateOrders)']].dtypes)
```

```python
# save the cleaned dataset to a new csv file
df.to_csv('../data/cleaned_supply_chain.csv', index=False)

print("Saved!")
```

```python
# Loading to sql database
from sqlalchemy import create_engine
engine = create_engine('sqlite:///../data/cleaned_supply_chain.db')
print("Database connected successfully!")
```

```python
# load the cleaned dataset into a new table in the database
df.to_sql('cleaned_supply_chain', con=engine, if_exists='replace', index=False)
print("Data loaded into database successfully!")
```

```python
test_query = "SELECT * FROM cleaned_supply_chain LIMIT 5;"
test_result = pd.read_sql_query(test_query, con=engine)
print(test_result)
```

```python
# delivery delay rate by shipping mode
query1 = """
SELECT 
    "Shipping Mode",
    count(*) AS total_orders,
    sum("Late_delivery_risk") AS late_deliveries,
    round(sum(Late_delivery_risk) * 100.0 / count(*), 2) AS delivery_delay_rate
FROM cleaned_supply_chain
GROUP BY "Shipping Mode"
ORDER BY delivery_delay_rate DESC;
    
"""
result1 = pd.read_sql_query(query1, con=engine)
print(result1)
```

```python
# revenue by market
query2 = """
SELECT
    "Market",
    round(sum("Sales"), 2) as total_sales,
    round(sum("Order Profit Per Order"), 2) as total_profit,
    count(*) as total_orders
FROM cleaned_supply_chain   
GROUP BY "Market"
ORDER BY total_sales DESC;
"""
result2 = pd.read_sql_query(query2, con=engine)
print(result2)
```

```python
# top 5 most profitable products categories
query3 = """
SELECT
    "Category Name",
    round(sum("Order Profit Per Order"), 2) as total_profit,
    count(*) as total_orders
FROM cleaned_supply_chain
GROUP BY "Category Name"
ORDER BY total_profit DESC
LIMIT 5;
"""
result3 = pd.read_sql_query(query3, con=engine)
print(result3)
```

```python
# monthly sales trend
query4 = """
SELECT  
    strftime('%Y-%m', "order date (DateOrders)") AS order_month,
    count(*) AS total_orders,
    round(sum("Sales"), 2) AS total_sales,
    round(sum("Order Profit Per Order"), 2) AS total_profit
FROM cleaned_supply_chain
GROUP BY order_month
ORDER BY order_month;
"""
result4 = pd.read_sql_query(query4, con=engine)
print(result4)
```
